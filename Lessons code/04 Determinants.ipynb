## 4.3 Cramer's rule

```python
X = Matrix([[3, -2, 6], [-5, 4, 8]])
X.rref()
```

```python
A = Matrix([[3, -2], [-5, 4]])
b = Matrix([6, 8])

# use cramer's rule to solve Ax = b
D = A.det()
A1 = b.row_join(A[:, 1])
D1 = A1.det()
A2 = A[:, 0].row_join(b)
D2 = A2.det()
x1 = D1 / D
x2 = D2 / D
x1, x2
```
