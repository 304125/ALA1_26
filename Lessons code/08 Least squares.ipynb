## 6.4 The Gram-Schmidt process

```python
x1 = Matrix([3, 6, 0])
x2 = Matrix([1, 2, 2])

# check for linear independence
A = Matrix.hstack(x1, x2)
A.rref()
```

```python
# check for orthogonality (dot product = 0)
x1.dot(x2)
```

```python
A.columnspace()
```

```python
GramSchmidt(A.columnspace())
```

```python
x1, x2 - x2.project(x1)
```

## 6.5 Least Squares Problems

```python
A = Matrix([[1, 2], [2, 1], [0, 1]])
b = Matrix([3, 3, 2])

# check if Ax = b has a solution
A.row_join(b).rref()
```

```python
ATA = A.T * A
ATB = A.T * b

M = ATA.row_join(ATB)
M = M.rref()
M
```

```python
# save the last column of M as x_hat
x_hat = M[0][:, -1]
x_hat
```

```python
b_hat = A*x_hat
b_hat
```

```python
b - b_hat
```

## 6.6 Machine Learning and Linear Models

```python
# B0 + B1 * ln(W) = p

W = [44, 61, 81, 113, 131]
p = Matrix([91, 98, 103, 110, 112])

# x has two columns: column of 1s and column of ln(W)
x = Matrix([[1, log(44)], [1, log(61)], [1, log(81)], [1, log(113)], [1, log(131)]])

XTX = x.T * x
XTP = x.T * p
M = XTX.row_join(XTP)
M = M.rref()
M
```

```python
# make last column entries in M into floats

M_float = M[0][:, -1].applyfunc(lambda x: x.evalf(5))
M_float
```

```python
B0 = M_float[0]
B1 = M_float[1]

print("p", "=", B0, "+", B1, "* ln(W)")
```
