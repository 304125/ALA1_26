## 5.1 Eigenvectors and Eigenvalues

```python
A = Matrix([[1, 6], [5, 2]])
u = Matrix([[-12], [10]])
v = Matrix([[3], [-2]])
A*u, (A*u).row_join(u).rref()
```

```python
A*v, (A*v).row_join(v).rref()
```

## 5.2 The characteristic equation

```python
A = Matrix([[2, 3], [3, -6]])
A.eigenvects()
```

```python
a = symbols('a')
A = Matrix([[2-a, 3], [3, -6-a]])
determinant = A.det()
determinant
```

```python
# solve for a
roots = solve(determinant, a)
roots
```

## 5.3 Diagonalization

```python
A = Matrix([[1-a, 3, 3], [-3, -5-a, -3], [3, 3, 1-a]])
solve(A.det(), a)
```

```python
A = Matrix([[1, 3, 3], [-3, -5, -3], [3, 3, 1]])
A.eigenvects()
```

```python
D = Matrix([[1, 0, 0], [0, -2, 0], [0, 0, -2]])
P = Matrix([[1, -1, -1], [-1, 1, 0], [1, 0, 1]])
P*D*P.inv()
```
