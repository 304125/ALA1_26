## SVD Example

```python
A = Matrix([[0, 1], [1, 0], [1, 1]])
AAT = A * A.T
AAT
```

```python
l = symbols('l')
char_eq = AAT.charpoly(l).as_expr()
char_eq
```

```python
roots = solve(char_eq, l)
roots
```

```python
AAT.eigenvects()
```

```python
M = AAT - roots[0]*Matrix.eye(3)
M.rref()
```

```python
M = AAT - roots[1]*Matrix.eye(3)
M.rref()
```

```python
M = AAT - roots[2]*Matrix.eye(3)
M.rref()
```

```python
u1 = AAT.eigenvects()[0][2][0]
u2 = AAT.eigenvects()[1][2][0]
u3 = AAT.eigenvects()[2][2][0]
u1, u2, u3
```

```python
n1 = u1.normalized()
n2 = u2.normalized()
n3 = u3.normalized()
n1, n2, n3
```

```python
U = Matrix.hstack(n3, n2, n1)
U
```

```python
l1 = roots[0]
l2 = roots[1]
l3 = roots[2]

sigma1 = sqrt(l1)
sigma2 = sqrt(l2)
sigma3 = sqrt(l3)
sigma1, sigma2, sigma3
```

```python
Sigma = Matrix([[sigma3, 0], [0, sigma2], [0, 0]])
Sigma
```

```python
v3 = 1/sigma3 * u3.T * A
v2 = 1/sigma2 * u2.T * A
v3, v2
```

```python
vn3 = v3.normalized()
vn2 = v2.normalized()
vn3, vn2
```

```python
V = Matrix.hstack(vn3.T, vn2.T)
V
```

```python
U*Sigma*V.T, A
```
