## 1.1 System of linear equations

```python
m = Matrix([[1, 2, 3, 4], [3, 6, 9, 12]])
m.rref()
```

```python
h, k = symbols('h k')
m = Matrix([[2, -1, h], [-6, 3, k]])
# reduce to echelon form but not row reduced echelon form
m.echelon_form()
```

## 1.4 Matrix equation

```python
m = Matrix([[1, 1, 1, 6], [2, 1, 3, 11], [1, 2, 1, 8]])
m.rref()
```

```python
# remove the last column from m
m2 = m[:, :-1]
# save last column of m.rref() as b
b = m.rref()[0].col(-1)
m2, b, m2*b
```
