## 2.1 Matrix operations

```python
Matrix([[0, 1, -1], [3, 0, 2]]) + Matrix([[1, 0, 0], [1, 1, 3]])
```

```python
3*Matrix([[0, 1, -1], [3, 0, 2]])
```

```python
A = Matrix([[1, 2], [2, 3], [3, 4]])
B = Matrix([[2, -1], [3, 1]])
A*B
```

```python
A = Matrix([[2, 1, 2], [4, 3, 2], [9, 7, 4]])
B = Matrix([[1, 1, 1, 2], [2, 1, 2, -1], [3, 4, 5, 4]])
A*B
```

```python
# remember indexing starts at 0!
a = A.row(1)
b = B.col(2)
a, b
```

```python
a*b
```

```python
A = Matrix([[4, 1], [3, 2], [7, 8]])
A, A.T
```

```python
A = Matrix([[1, -3], [-2, 4]])
x = Matrix([5, 3])
(A*x).T, x*x.T, x.T*x
```

## 2.2 The inverse of a matrix

```python
A = Matrix([[1, 2], [3, 4]])
A.det()
```

```python
A*A.inv()
```

```python
A = Matrix([[1, 2], [3, 4]])
b = Matrix([5, 6])
A.inv()*b
```

```python
A = Matrix([[1, 2, 1], [4, 5, 6], [0, 0, 2]])
I = Matrix.eye(3)
A.row_join(I).rref()
```

```python
A.inv()
```

## 2.5 Matrix factorization

```python
A = Matrix([[2, 4, -1, 5, -2], [-4, -5, 3, -8, 1], [2, -5, -4, 1, 8], [-6, 0, 7, -3, 1]])
L, U, piv = A.LUdecomposition()
L, U
```

```python
L*U
```

```python
b = Matrix([2, -4, 2, -6])
A.row_join(b).rref()
```
