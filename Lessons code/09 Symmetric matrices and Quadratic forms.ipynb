# 7.1 Diagonalization of Symmetric Matrices

```python
P, D = A.diagonalize()
P, D
```

```python
P.inv()
```

```python
P * D * P.inv(), A
```

### Orthogonal Diagonalization

```python
A = Matrix([[3, -2, 4], [-2, 6, 2], [4, 2, 3]])
eigenvectors = A.eigenvects()
eigenvectors
```

```python
z1 = eigenvectors[1][2][1]
z2 = eigenvectors[1][2][0]
z3 = eigenvectors[0][2][0]
z1, z2, z3
```

```python
u1, u2 = GramSchmidt((z1, z2))
u1, u2
```

```python
n1 = u1.normalized()
n2 = u2.normalized()
n3 = z3.normalized()
n1, n2, n3
```

```python
P = Matrix.hstack(n1, n2, n3)
P.T, P
```

```python
D = Matrix([[7, 0, 0], [0, 7, 0], [0, 0, -2]])
P * D * P.T, A
```

# 7.2 Quadratic Forms

### Quadratic form (diagonal matrix)

```python
x1, x2 = symbols('x1 x2')
A = Matrix([[4, 0], [0, 3]])
x = Matrix([x1, x2])
Q = x.T * A * x
Q = Q[0]
Q
```

### Quadratic form (symmetric matrix)

```python
A = Matrix([[3, -2], [-2, 7]])
x = Matrix([x1, x2])
Q = x.T * A * x
Q = Q[0]
Q
```

### Compute Q(x) for a given x

```python
Qx = x1**2 - 8*x1*x2 - 5*x2**2
x = Matrix([-3, 1])
Qx.subs({x1: x[0], x2: x[1]})
```

```python
x = Matrix([2, -2])
Qx.subs({x1: x[0], x2: x[1]})
```

```python
x = Matrix([1, -3])
Qx.subs({x1: x[0], x2: x[1]})
```

### Change of variables to simplify Q

```python
A = Matrix([[1, -4], [-4, -5]])
A.eigenvects()
```

```python
lambda_1 = A.eigenvects()[0][0]
lambda_2 = A.eigenvects()[1][0]
lambda_1, lambda_2
```

```python
u1 = A.eigenvects()[0][2][0].normalized()
u2 = A.eigenvects()[1][2][0].normalized()
u1, u2
```

```python
P = Matrix.hstack(u2, u1)
D = Matrix([[lambda_2, 0], [0, lambda_1]])
P, D
```

```python
x = Matrix([2, -2])
y = P.T * x
y
```

```python
Q = x.T * A * x
Q = Q[0]
Q
```

```python
Q_y = y.T * D * y
Q_y = Q_y[0]
Q_y
```

### Finding axes for change of variable

```python
Qx = 5*x1**2 - 4*x1*x2 + 5*x2**2
A = Matrix([[5, -2], [-2, 5]])
```

```python
A.eigenvects()
```

```python
lambda_1 = A.eigenvects()[0][0]
lambda_2 = A.eigenvects()[1][0]
lambda_1, lambda_2
```

```python
u1 = A.eigenvects()[0][2][0].normalized()
u2 = A.eigenvects()[1][2][0].normalized()
u1, u2
```

```python
P = Matrix.hstack(u1, u2)
D = Matrix([[lambda_1, 0], [0, lambda_2]])
P, D
```

```python
y1, y2 = symbols('y1 y2')
y = Matrix([y1, y2])
Q_y = y.T * D * y
Q_y = Q_y[0]
Q_y
```
