## 1.7 Linear Dependence

```python
v1 = Matrix([1, 2, 3])
v2 = Matrix([4, 5, 6])
v3 = Matrix([2, 1, 0])
# concatenate vectors into a matrix
m = Matrix.hstack(v1, v2, v3)
m.rref()
```

## 1.8 Linear Transformations

```python
A = Matrix([[1, -3], [3, 5], [-1, 7]])
u = Matrix([2, -1])
b = Matrix([3, 2, -5])
c = Matrix([3, 2, 5])
```

```python
A*u
```

```python
# concatinate A and b into a matrix using hstack
A.row_join(b).rref()
```

```python
A.row_join(c).echelon_form()
```

## 1.9 Matrix of Linear Transformations

```python
m = Matrix([[1, -4, 8, 1], [0, 2, -1, 3], [0, 0, 0, 5]])
m.rref()
```
