## 4.1 Vector Spaces and Subspaces

```python
n = symbols('n')
v1 = Matrix([1, -1, -2])
v2 = Matrix([5, -4, -7])
v3 = Matrix([-3, 1, 0])
y = Matrix([-4, 3, n])
A = Matrix.hstack(v1, v2, v3, y)
A.echelon_form()
```

## 4.2 Null Spaces, Column Spaces, Row Spaces, and Linear Transformations

```python
A = Matrix([[1, -3, -2], [-5, 9, 1]])
A.rref()
```

```python
u = Matrix([5, 3, -2])
A*u
```

```python
A = Matrix([[2, 2, 1], [4, 1, 0]])
v1 = Matrix([1, 1, 0])
v2 = Matrix([1, -4, 6])
A*v1, A*v2
```

```python
A.rref()
```

```python
A = Matrix([[-3, 6, -1, 1, -7], [1, -2, 2, 3, -1], [2, -4, 5, 8, -4]])
A.rref()
```

```python
A.nullspace()
```

```python
A = Matrix([[2, 4, -2, 1], [-2, -5, 7, 3], [3, 7, -8, 6]])
A.nullspace()
```

```python
u = Matrix([3, -2, -1, 0])
v = Matrix([3, -1, 3])
A*u, A.row_join(v).rref()
```

## 4.4 Coordinate Systems

```python
v1 = Matrix([1, 0])
v2 = Matrix([1, 1])
3*v1 + 4*v2
```

```python
v1 = Matrix([2, 1])
v2 = Matrix([1, 2])
x_b = Matrix([5, -2])
x = x_b[0]*v1 + x_b[1]*v2
x
```
