## 6.1 Inner Product, Length and Orthogonality

```python
v = Matrix([1, -2, 2, 0])
v_unit = v / sqrt(v.dot(v))
v_unit
```

## 6.2 Orthogonal Sets

```python
u1 = Matrix([3, 1, 1])
u2 = Matrix([-1, 2, 1])
u3 = Matrix([-1, -4, 7])
y = Matrix([6, 1, -8])
U = Matrix.hstack(u1, u2, u3)
U.T*U
```

```python
c1 = y.dot(u1) / u1.dot(u1)
c2 = y.dot(u2) / u2.dot(u2)
c3 = y.dot(u3) / u3.dot(u3)
c1, c2, c3
```

```python
y = Matrix([3, R(1,2), -4])
c1 = y.dot(u1) / u1.dot(u1)
c2 = y.dot(u2) / u2.dot(u2)
c3 = y.dot(u3) / u3.dot(u3)
c1, c2, c3
```

## 6.3 Orthogonal Projections

```python
y_bar = Matrix([7, 6])
u_bar = Matrix([4, 2])
y_hat = y_bar.dot(u_bar) / u_bar.dot(u_bar) * u_bar
y_hat
```

```python
z_bar = y_bar - y_hat
z_bar
```

```python
# check
y_hat.dot(z_bar)
```

```python
u1 = Matrix([2, 5, -1])
u2 = Matrix([-2, 1, 1])
y_bar = Matrix([1, 2, 3])

y_hat = y_bar.dot(u1) / u1.dot(u1) * u1 + y_bar.dot(u2) / u2.dot(u2) * u2
y_hat
```

```python
z_bar = y_bar - y_hat
z_bar
```

```python
# check
y_hat.dot(z_bar)
```

```python
# make y_hat and z_bar orthonormal
y_hat_unit = y_hat / sqrt(y_hat.dot(y_hat))
z_bar_unit = z_bar / sqrt(z_bar.dot(z_bar))
y_hat_unit, z_bar_unit
```
